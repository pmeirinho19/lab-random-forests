```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import GridSearchCV
from sklearn.feature_selection import VarianceThreshold
```

```python
categorical = pd.read_csv('files_for_lab/categorical.csv')
numerical = pd.read_csv('files_for_lab/numerical.csv')
target = pd.read_csv('files_for_lab/target.csv')
```

```python
#Features automated selection - numericals

transformer = MinMaxScaler()
transformer.fit(numerical)

num_scaled = transformer.transform(numerical)
numerical_scaled = pd.DataFrame(num_scaled, columns = numerical.columns)
numerical_scaled



```

```python
var_threshold = 0.02

sel = VarianceThreshold(threshold=(var_threshold))

sel = sel.fit(numerical_scaled)
temp = sel.transform(numerical_scaled)
temp = pd.DataFrame(data = temp, columns = list(sel.get_feature_names_out()))

print(temp.shape)
```

```python
cols = list(categorical.columns)

for cols in categorical:
    print(categorical[cols].value_counts())
```

```python
### we have to drop some categorical columns because when we use OneHot Encoder
    #we are getting to much columns(315)

#Columns to drop since they have to much values to encode
    #CLUSTER
    #DOMAIN_B
    #ODATEW_YR
    #ODATEW_MM
    #DOB_YR
    #DOB_MM
    #MINRDATE_YR
    #MINRDATE_MM
    #MAXRDATE_YR
    #LASTDATE_YR
    #LASTDATE_MM
    #FIRSTDATE_YR
    #FIRSTDATE_MM

cat_to_drop = ['CLUSTER','DOMAIN_B','ODATEW_YR','ODATEW_MM','DOB_YR','DOB_MM','MINRDATE_YR',\
 'MINRDATE_MM','MAXRDATE_YR','LASTDATE_YR','LASTDATE_MM','FIRSTDATE_YR','FIRSTDATE_MM']

len(cat_to_drop)

categorical.drop(columns = cat_to_drop, inplace=True)

categorical.shape
```

```python
df = pd.concat([temp,categorical,target],axis = 1)
df
```
